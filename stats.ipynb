```python
import pandas as pd
import numpy as np

df = pd.read_csv("rural_india_corpus.csv", sep=";")
df.head()
df.head(50).to_csv("sample.csv", sep=";", index=False)

df.head()
```

```python
print("Total number of articles: ", len(df))
print("\nNumber of articles per language: ")
print(df[df.columns[2:]].count())
```

```python
pd.options.mode.chained_assignment = None  # disable SettingWithCopyWarning warning

df_asint = df.loc[:, 'English':].notnull().astype('int')
cotrans = df_asint.T.dot(df_asint)
for i, lang in enumerate(cotrans.columns):
    cotrans[lang].iloc[i] = 0
cotrans
```

```python
all_topics = set('')
index_by_topic = []
df['topics'].fillna('', inplace=True)
for topic in df['topics'].values:
    index_by_topic.append(np.array([topic.strip() for topic in topic.split(',') if topic != '']))
    all_topics.update(topic.split(','))

all_topics = [topic.strip() for topic in all_topics if topic != '']
print(all_topics)
```

```python
df[all_topics] = 0
for i, topics in enumerate(index_by_topic):
    for t in topics:
        df.at[i, t] = 1
```

```python
df_tmp = df.loc[:, 'Little Takes':]
cotopic = df_tmp.T.dot(df_tmp)
for i, lang in enumerate(cotopic.columns):
    cotopic[lang].iloc[i] = 0
cotopic
```
